```python
def make_alternating(text):
    
return_text = ""
for i in range(len(text)):
    word = text[i]
    if word.isalpha():
        if i%2 == 0:
            return_text += word.lower()
        else:
            return_text += word.upper()

    else:
        return_text += word
return return_text


test_string = input("Enter a string: ")
result = make_alternating(test_string)
print(result)
```

```python
# Exercise 7: Simple Grid Game
# Write your code here
# Creating a 2D list (8x8 grid)
SIZE = 8
grid = [['.' for col in range(SIZE)] for row in range(SIZE)]


player_row = 4
player_col = 4
game_status = True
move_number = 1


def gamesetup():
    grid[4][4] = 'P'
    for row in grid:
        print(' '.join(row))
    return

gamesetup()


while game_status == True:
    grid[player_row][player_col] = '-'
row_old = player_row
col_old = player_col


print ("Move number: ", move_number)
move = input ("Move (u/d/l/r/q):")


if move == "u":
    player_row -=1
elif move == "d":
    player_row += 1
elif move == "r":
    player_col +=1
elif move == "l":
    player_col -=1
elif move == "q":
    game_status = False
move = ""


grid[player_row][player_col] = 'P'
for row in grid:
    print(' '.join(row))
#grid[player_row][player_col] = '.'
grid[row_old][col_old] = '.'
move_number +=1
```

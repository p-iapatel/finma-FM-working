```python
prices = {
    "box_of_spaghetti": 4,
    "lasagna": 5,
    "hamburger": 2
}

quantity = {
    "box_of_spaghetti": 6,
    "lasagna": 10,
    "hamburger": 0
}
money_spent=0
for item, price in prices.items():
    if price >= 5:
        money_spent += price * quantity[item]
        print(f"Jan spent {money_spent} $ on products costing $5 or more.")
```

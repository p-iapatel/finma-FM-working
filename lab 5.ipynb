```python
#exercise 6

#portfolios
investor1=["AAPL", "GOOGL", "MSFT", "TSLA", "NVDA"]
investor2=["AAPL", "AMZN", "MSFT", "META", "NFLX"]

common_stocks = set(investor1).intersection(investor2)
print("Stocks both investors own are",common_stocks)

all_stocks = set(investor1).union(investor2)
print("All unique stocks across both portfolios are", all_stocks )

unique_to_investor1 = set(investor1).difference(investor2)
print("Stocks that only Investor 1 hold are", unique_to_investor1 )

exclusive_stocks = set(investor1).symmetric_difference(investor2)
print("Stocks that only one investor own are", exclusive_stocks)
```

```python
#exercise 7

stock_database = {
    "AAPL" : ("Apple Inc.", "Technology", 255.52),
    "JPM" : ("JPMorgan Chase", "Finance", 312.47),
    "XOM" : ("Exxon Mobil", "Energy", 129.89),
    "AMZN" : ("Amazon.com Inc.", "Consumer Discretionary", 239.12),
    'MSFT' : ("Microsoft Corp.", "Technology", 459.86)
}

for symbol, (name, sector, price) in stock_database.items():
    prices = [price for name, sector, price in stock_database.values()]
    if sector == "Technology":
        print(f"{name} is in the {sector} sector.")
print("The most expensive stock is worth", max(prices), "USD.")
print("The average price of all stocks is", sum(prices)/len(prices), "USD.")

```

```python
#exercise 1
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def fahrenheit_to_celsius(fahrenheit):
    return (fahrenheit - 32) * 5/9

print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(100))
print(fahrenheit_to_celsius(32))
print(fahrenheit_to_celsius(212))
```

```python
#exercise 4
def calculate_portfolio_return(initial_value, final_value, years=1):
    return (((final_value/initial_value)**(1/years)-1) * 100)

#1 year return (using default)
print(calculate_portfolio_return(10000,11000))

#3 year return
print(calculate_portfolio_return(10000,13000,3))

    
```

```python
#exercise 6
def summarise_positions(*positions):
    total = sum(positions)
    i = 1

    for value in positions:
        percentage = (value/total) * 100
        print(f"position {i}: {value}, {percentage}%")
        i += 1
     
    return total

print("Total:", summarise_positions(15000, 8000, 12000, 5000))
```
